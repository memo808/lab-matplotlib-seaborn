## Coding Session I - 30 minutes

From the Fitbit data, we want to visually understand: 

1. **How the average number of steps change by month. Use the appropriate visualization to show the median steps by month?**
1. **Is Fitbitter more active on weekend or workdays?**

##### All plots must be in the same jupyter notebook cell.

Hints:

- Use `Months_encoded` and `Week_or Weekend` columns
- Use `matplolib.pyplot` object oriented API
- Set your size figure to `12,4`
- Explore `plt.sca`
- Explore `plt.xticks`
- Save your figures

```python
# Driver 1 enter codes here
media_pasos_mes=fitbit.groupby('Months_encoded',as_index=False).agg({'Steps':'mean'})
mediana_pasos_mes=fitbit.groupby('Months_encoded',as_index=False).agg({'Steps':'median'})
media_pasos_dia=fitbit.groupby('Work_or_Weekend',as_index=False).agg({'Steps':'mean'})
mediana_pasos_dia=fitbit.groupby('Work_or_Weekend',as_index=False).agg({'Steps':'median'})


fig,(x1,x2,x3)=plt.subplots(1,3,figsize=(12,4))

x1.bar(media_pasos_mes['Months_encoded'],media_pasos_mes['Steps'],tick_label=['January','February','March','April','May','June','July','August','September','October','November','December'])
x2.bar(mediana_pasos_mes['Months_encoded'],mediana_pasos_mes['Steps'],tick_label=['January','February','March','April','May','June','July','August','September','October','November','December'])

x3.bar(media_pasos_dia['Work_or_Weekend'],media_pasos_dia['Steps'],tick_label=['Weekend','Workdays'])

x1.set_xticklabels(['January','February','March','April','May','June','July','August','September','October','November','December'],rotation=90)
x2.set_xticklabels(['January','February','March','April','May','June','July','August','September','October','November','December'],rotation=90)
x1.set_title('Month vs Steps (mean)',size=12)
x2.set_title('Month vs Steps (median)',size=12)
x3.set_title('Average steps weekend vs average steps workdays',size=8)
x1.set_ylabel('Steps',size=10)
```

```python
#1. Se puede ver tanto en la media como en la mediana que el mes con mayores pasos es abril, seguido de junio y en tercer lugar agosto.
#2. Se puede observar en el tercer grÃ¡fico que en promedio se hacen mÃ¡s pasos en dÃ­as de trabajo que en los fines de semana.
```
